```python
#import pandas
import pandas as pd
```

```python
# Get football data from ESPN.
football_df = pd.read_html("http://www.espn.com/nfl/superbowl/history/winners", header=1)
```

```python
# Get dataFrame
football_df = football_df[0]
football_df
```

```python
# Get year of each championship.
football_df[["DAY", "YEAR"]] = football_df["DATE"].str.split(",", expand=True)
football_df
```

```python
# get rid of scores so teams can be split into Winner and Loser. 
import re
football_df["RESULT"].replace({r"\d": ""}, inplace= True, regex= True)
football_df
```

```python
#Use split to separate winner and loser into columns.
football_df[["WINNER", "LOSER"]] = football_df["RESULT"].str.split(",", expand=True)
football_df
```

```python
# Get city name only for winners using regex.
football_df["WINNER"].replace({r"[RGJ].{2,}[s]": ""}, inplace= True, regex= True)
football_df
```

```python
# Get city name only for losers.
football_df["LOSER"].replace({r"[RG].{2,}[s]": ""}, inplace= True, regex= True)
football_df
```

```python
# Create new dataframe with only year, winner, and loser info.
football_df2 = football_df[["YEAR", "WINNER", "LOSER"]]
football_df2
```

```python
# Change New England to Boston and Carolina to Charlotte.
football_df2.loc[football_df2["WINNER"] == "New England ", "WINNER"] = "Boston"
football_df2.loc[football_df2["LOSER"] == " New England ", "LOSER"] = "Boston"
football_df2.loc[football_df2["LOSER"] == " Carolina ", "LOSER"] = "Charlotte"
football_df2.loc[football_df2["LOSER"] == " Arizona ", "LOSER"] = "Phoenix"
football_df2.loc[football_df2["LOSER"] == " Tennessee ", "LOSER"] = "Nashville"

football_df2
```

```python
# Get baseball information.
baseball_df = pd.read_html("https://www.baseball-reference.com/postseason/world-series.shtml")
baseball_df = baseball_df[0]
baseball_df
```

```python
# Create function to determine who won the WS.
def winners(row):
    if row["Wins"] < row["Wins.1"]:
        return row["NL Winner"]
    elif row["Wins"] > row["Wins.1"]:
        return row["AL Winner"]
    
```

```python
# Create function to determine who lost the WS.

def losers(row):
    if row["Wins"] < row["Wins.1"]:
        return row["AL Winner"]
    elif row["Wins"] > row["Wins.1"]:
        return row["NL Winner"]
```

```python
# Apply function for Winners
baseball_df["WINNER"] = baseball_df.apply(lambda row: winners(row), axis=1)
```

```python
# Apply function for Losers.
baseball_df["LOSER"] = baseball_df.apply(lambda row: losers(row), axis=1)
```

```python
# Capitalize Year column
baseball_df = baseball_df.rename(columns={"Year": "YEAR"})
```

```python
# Create new dataframe with year, winner, and loser only.
baseball_df2 = baseball_df[["YEAR", "WINNER", "LOSER"]]
baseball_df2
```

```python
pd.set_option("display.max_rows", None)
baseball_df2
```

```python
# Get city name for winners only using regex.
baseball_df2["WINNER"].replace({r"[A-Z]\w{2,}[s]$": ""}, inplace= True, regex= True)
baseball_df2["WINNER"].replace({r"[RW].{2,}[x]$": ""}, inplace= True, regex= True)
baseball_df2["WINNER"].replace({r"[B]\w{2}[e]\s$": ""}, inplace= True, regex= True)
baseball_df2
```

```python
# Get city name for losers only using regex.
baseball_df2["LOSER"].replace({r"[A-Z]\w{2,}[s]$": ""}, inplace= True, regex= True)
baseball_df2["LOSER"].replace({r"[RW].{2,}[x]$": ""}, inplace= True, regex= True)
baseball_df2
```

```python
# Change state to city name for merge.
baseball_df2.loc[baseball_df2["WINNER"] == "Minnesota ", "WINNER"] = "Minneapolis"
baseball_df2.loc[baseball_df2["LOSER"] == "Minnesota ", "LOSER"] = "Minneapolis"
baseball_df2.loc[baseball_df2["WINNER"] == "Texas ", "WINNER"] = "Dallas"
baseball_df2.loc[baseball_df2["LOSER"] == "Texas ", "LOSER"] = "Dallas"
baseball_df2.loc[baseball_df2["WINNER"] == "Florida ", "WINNER"] = "Miami"
baseball_df2.loc[baseball_df2["LOSER"] == "Florida ", "LOSER"] = "Miami"
baseball_df2.loc[baseball_df2["WINNER"] == "Arizona ", "WINNER"] = "Phoenix"
baseball_df2
```

```python
# Get basketball information from wikipedia and amount of tables
basketball_df = pd.read_html("https://en.wikipedia.org/wiki/List_of_NBA_champions")
len(basketball_df)
```

```python
# create dataframe
basketball_df = basketball_df[1]
basketball_df
```

```python
# Drop rows that did not have information
basketball_df = basketball_df.drop([0,4])
basketball_df
```

```python
# Split result into games won by East and West side.
basketball_df[["WEST", "EAST"]] = basketball_df["Result"].str.split("", expand=True)
basketball_df
```

```python
# Create function to determine winner. 
def winners(row):
    if row["WEST"] < row["EAST"]:
        return row["Eastern champions"]
    elif row["WEST"] > row["EAST"]:
        return row["Western champions"]
    
```

```python
# Create function to determine loser. 

def losers(row):
    if row["WEST"] < row["EAST"]:
        return row["Western champions"]
    elif row["WEST"] > row["EAST"]:
        return row["Eastern champions"]
```

```python
# Apply function
basketball_df["Winner"] = basketball_df.apply(lambda row: winners(row), axis=1)
```

```python
# Apply function
basketball_df["Loser"] = basketball_df.apply(lambda row: losers(row), axis=1)
```

```python
# Check dataframe
basketball_df
```

```python
# Split winner and losers into only city and team names.
basketball_df[["WINNER", "EXTRA"]] = basketball_df["Winner"].str.split("(", n=1, expand=True)
basketball_df[["LOSER", "EXTRA2"]] = basketball_df["Loser"].str.split("(", n=1, expand=True)
basketball_df
```

```python
# Capitalize year
basketball_df = basketball_df.rename(columns={"Year":"YEAR"})
basketball_df
```

```python
# Create new dataframe with only year, winner, and loser
basketball_df2 = basketball_df[["YEAR", "WINNER", "LOSER"]]
basketball_df2
```

```python
# Get city for winner using regex
basketball_df2["WINNER"].replace({r"[A-Z0-9].{1}\w{1,}[st]\s$": ""}, inplace= True, regex= True)
basketball_df2["WINNER"].replace({r"[T].{1,}[l]\s$": ""}, inplace= True, regex= True)
basketball_df2
```

```python
# Get city for loser using regex
basketball_df2["LOSER"].replace({r"[A-Z0-9].{1}\w{1,}[stczr]\s$": ""}, inplace= True, regex= True)
basketball_df2["LOSER"].replace({r"[T].{1,}[l]\s$": ""}, inplace= True, regex= True)
basketball_df2
```

```python
# Change metro areas or states to cities.
basketball_df2.loc[basketball_df2["WINNER"] == "Golden State ", "WINNER"] = "San Francisco"
basketball_df2.loc[basketball_df2["LOSER"] == "Golden State ", "LOSER"] = "San Francisco"
basketball_df2.loc[basketball_df2["LOSER"] == "Indiana ", "LOSER"] = "Indianapolis"
basketball_df2
```

```python
# Create a league column on each dataset
basketball_df2["LEAGUE"] = "NBA"
baseball_df2["LEAGUE"] = "MLB"
hockey_df2["LEAGUE"] = "NHL"
football_df2["LEAGUE"] = "NFL"
```

```python
basketball_df2
```

```python
# Clean up year on basketball before merge
basketball_df2[["YEAR", "EXTRA"]] = basketball_df2["YEAR"].str.split("[", n=1, expand=True) 
basketball_df2 = basketball_df2.drop(columns=["EXTRA"])
basketball_df2
```

```python
# Merge datasets
merged_df = pd.concat([baseball_df2, basketball_df2, hockey_df2, football_df2])
merged_df
```

```python
merged_df.set_index("YEAR")
```

```python
winner = merged_df["WINNER"].str.strip()
loser = merged_df["LOSER"].str.strip()
```

```python
wins = winner.value_counts()
losses = loser.value_counts()
print(wins)
print(losses)
```

```python
appearances = {"wins" : wins, "losses": losses}
appearances_df = pd.DataFrame(appearances)
appearances_df
```

```python
appearances_df = appearances_df.fillna(0)
appearances_df
```

```python
appearances_df["Total"] = appearances_df["wins"] + appearances_df["losses"]
appearances_df
```

```python
appearances_df["Winning Percentage"] = appearances_df.loc[appearances_df["wins"] == 0, "Winning Percentage"] = 0
appearances_df["Winning Percentage"] = appearances_df.loc[(appearances_df["wins"] > 0) & (appearances_df["losses"] == 0), "Winning Percentage"] = 1
appearances_df["Winning Percentage"] = appearances_df.loc[appearances_df["wins"] > 0, "Winning Percentage"] = appearances_df["wins"] / appearances_df["Total"]
appearances_df
```

```python
appearances_df.index.rename("city", inplace=True)
appearances_df
```

```python
# Create csv from dataframe
merged_df.to_csv("champions.csv")
appearances_df.to_csv("appearances.csv")
```
