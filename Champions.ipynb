```python
# Create function to remove whitespace
def whitespace_remover(dataframe):
   
    # iterating over the columns
    for i in dataframe.columns:
         
        # checking datatype of each columns
        if dataframe[i].dtype == 'object':
             
            # applying strip function on column
            dataframe[i] = dataframe[i].map(str.strip)
        else:
             
            # if condn. is False then it will do nothing.
            pass
```

```python
# Get baseball information.
baseball_df = pd.read_html("https://www.baseball-reference.com/postseason/world-series.shtml")
baseball_df = baseball_df[0]
baseball_df
```

```python
# Create function to determine who won the WS.
def winners(row):
    if row["Wins"] < row["Wins.1"]:
        return row["NL Winner"]
    elif row["Wins"] > row["Wins.1"]:
        return row["AL Winner"]
    
```

```python
# Create function to determine who lost the WS.

def losers(row):
    if row["Wins"] < row["Wins.1"]:
        return row["AL Winner"]
    elif row["Wins"] > row["Wins.1"]:
        return row["NL Winner"]
```

```python
# Apply function for Winners
baseball_df["WINNER"] = baseball_df.apply(lambda row: winners(row), axis=1)
```

```python
# Apply function for Losers.
baseball_df["LOSER"] = baseball_df.apply(lambda row: losers(row), axis=1)
```

```python
# Create function to determine winner. 
def winners(row):
    if row["WEST"] < row["EAST"]:
        return row["Eastern champions"]
    elif row["WEST"] > row["EAST"]:
        return row["Western champions"]
    
```

```python
# Create function to determine loser. 

def losers(row):
    if row["WEST"] < row["EAST"]:
        return row["Western champions"]
    elif row["WEST"] > row["EAST"]:
        return row["Eastern champions"]
```

```python
# Apply function
basketball_df["Winner"] = basketball_df.apply(lambda row: winners(row), axis=1)
```

```python
# Create function to get championship appearances per sport for either winner or losing.
def count_values(df, col, dictionary):
    count = 0
    for city in city_list:
        if city in df[col].values:
            count = df[col].value_counts()[city]
            dictionary[city] = count
        else:
            dictionary[city] = 0
    return dictionary
```

```python
# Create a function that get the years values and put them a dictionary.
def league_teams(league, winner):
    dictionary = {}
    league_dict = {}
    if winner == True:
        league_df = championship_years_df.loc[championship_years_df["LEAGUE"] == league]
        league_dict = pd.Series(league_df.YEAR.values, index=league_df.WINNER).to_dict()
        league_dict = {k.strip(): v
                      for k, v in league_dict.items()}
        league_dict = {k: v.replace(" ", ",")
                      for k, v in league_dict.items()}
        for city in city_list:
            if city in league_dict.keys():
                dictionary[city] = league_dict.get(city)
            elif city not in league_dict.keys():
                dictionary[city] = "None"
        return dictionary
    if winner == False:
        league_df = losing_years_df.loc[losing_years_df["LEAGUE"] == league]
        league_dict = pd.Series(league_df.YEAR.values, index=league_df.LOSER).to_dict()
        league_dict = {k.strip(): v
                      for k, v in league_dict.items()}
        league_dict = {k: v.replace(" ", ",")
                      for k, v in league_dict.items()}
        for city in city_list:
            if city in league_dict.keys():
                dictionary[city] = league_dict.get(city)
            elif city not in league_dict.keys():
                dictionary[city] = "None"
        return dictionary
```
